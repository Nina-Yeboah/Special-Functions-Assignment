                                SPECIAL METHODS IN PROGRAMMING

```python
# Constructor
class Person:
    def __init__(self, name):
        self.name = name
user = Person("Nina")
print(user.name)
```

```python
# String representation
class Cat:
    def __str__(self):
        return "I am a cat!"
my_cat = Cat()
print(my_cat)
```

```python
#Length
class Group:
    def __len__(self):
        return 5
team = Group()
print(len(team))
```

```python
#Equality
class Box:
    def __init__(self, size):
        self.size = size
    def __eq__(self, other):
        return self.size == other.size
box1 = Box(10)
box2 = Box(10)
print(box1 == box2)
```

```python
#Operator overloading
class Score:
    def __init__(self, points):
        self.points = points
    def __add__(self, other):
        return Score(self.points + other.points)
player1 = Score(100)
player2 = Score(50)
total = player1 + player2
print(total.points)
```

```python
#Item access
class SecretAgent:
    def __getitem__(self, index):
        return "Classified Information"
agent = SecretAgent()
print(agent[0])
```

```python
#Item assignment
class SimpleStorage:
    def __init__(self):
        self.data = {}
    def __setitem__(self, key, value):
        self.data[key] = value
store = SimpleStorage()
store["power"] = 9000
print(store.data)
```

```python
#Membership testing
class Club:
    def __contains__(self, name):
        return name == "VIP"
guest_list = Club()
print("VIP" in guest_list)
print("Guest" in guest_list)
```

```python
#Callable objects
class Greeter:
    def __call__(self, name):
        return f"Hello, {name}!"
say_hi = Greeter()
print(say_hi("Nina"))
```

```python
#Iteration support
class Countdown:
    def __iter__(self):
        return iter([3, 2, 1])
timer = Countdown()
for number in timer:
    print(number)
```

```python
#Boolean evaluation
class Switch:
    def __init__(self, status):
        self.status = status
    def __bool__(self):
        return self.status == "ON"
light = Switch("ON")
if light:
    print("The light is shining!")
```

```python
#Destructor
class Temporary:
    def __del__(self):
        print("Object is destroyed!")
temp_obj = Temporary()
del temp_obj
```
